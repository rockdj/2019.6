```python
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
 
model = load_model('noise.h5')

for i in range(0,502):
    file_path= 'F:/Jupyter/2019.5/welding line/data_set_mini_mix/validation/pos/p'+ str(i)+'.jpg' ##图片地址
    #print(file_path)
    img=image.load_img(file_path,target_size=(384,192))
    x=image.img_to_array(img)
    x=np.expand_dims(x,axis=0)
    y=model.predict(x)
    if y == 1: 
        print(str(i)+': Positive')
    if y == 0:
        print(str(i)+':Negetive')
```
